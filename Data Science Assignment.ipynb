# 3.Model Building

```python
df.columns
```

```python
df['margin'] = df['item_price'] - df['cost_price']
```

```python
df['price_per_customer'] = df['item_price'] / df['footfall']
```

```python
X=df[['day_type', 'weather', 'location_type', 'time_of_day', 'item_type','footfall', 'item_price', 'cost_price']]
y=df['demand']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
r2 = r2_score(y_test, y_pred)
print("R2 Score:", r2)
```

```python
day_type = input("Day Type (Weekday/Weekend): ")
weather = input("Weather (Sunny/Cloudy/Rainy): ")
location_type = input("Location (College/Residential/Office): ")
time_of_day = input("Time (Morning/Afternoon/Evening): ")
item_type = input("Item (Maggie/Sandwich/Tea/Cold Drink): ")

footfall = int(input("Footfall: "))
cost_price = float(input("Cost Price: "))


day = d1[day_type]
weather = w[weather]
location = loc[location_type]
daytime = t[time_of_day]
item = menu[item_type]
```

```python
results = [] 
for price in range(20, 120, 10): # selling price # 20, 30, 40 ... 110
    test = [[day, weather, location, daytime, item,footfall, price, cost_price]]
    predicted_demand = int(model.predict(test)[0])
    profit = (price - cost_price) * predicted_demand
    results.append((price, predicted_demand, profit))
best = max(results, key=lambda x: x[2])  
print(best)
```

```python
print("\nBest Price Strategy:")
print("Price:", best[0])
print("Expected Demand:", int(best[1]))
print("Expected Profit:", int(best[2]),"Rs")
```

```python
print("\nAll Price Options:")
for r in results:
    print("Price:", r[0], "Demand:", r[1], "Profit:", r[2])
```
